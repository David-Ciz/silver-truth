```python
import tifffile
import matplotlib.pyplot as plt
import numpy as np
```

```python
marker = tifffile.imread('inputs-2020-07/BF-C2DL-HSC/01_GT/TRA/man_track0314.tif')
```

```python
def show_image(img, name):
    # Create a colormap that will work well for multiple values
# Using 'tab20' colormap which has 20 distinct colors
    plt.figure(figsize=(8, 8))
    plt.imshow(img, cmap='tab20')
    plt.colorbar(label='Square Values')
    plt.title(f'{name}')
    plt.show()
```

```python
show_image(marker, 'marker')
```

```python
np.unique(marker, return_counts=True)
```

```python
gt_image = tifffile.imread('inputs-2020-07/BF-C2DL-HSC/01_GT/SEG/man_seg0058.tif')
```

```python
show_image(gt_image, 'gt_image')
```
